We will now classify fighters:
Strikers = Fighters with Above average striking accuracy and defence
Grapplers = Fighters with Above average takedown defence and accuracy

```python
# Calculate average value for classification

avg_striking_acc = df_cleaned['significant_striking_accuracy'].mean()
avg_striking_def = df_cleaned['significant_strike_defence'].mean()
avg_takedown_acc = df_cleaned['takedown_accuracy'].mean()
avg_takedown_def = df_cleaned['takedown_defense'].mean()

# Function to classify Fighers
def classify_fighter(row):
    if row['significant_striking_accuracy'] > avg_striking_acc and row['significant_strike_defence'] > avg_striking_def:
        return 'Strikers'
    elif row['takedown_accuracy'] > avg_takedown_acc and row['takedown_defense'] > avg_takedown_def:
        return 'Grapplers'
    else:
        return 'Mixed Style'

# Apply the function to create the 'Fighting_Style' column
df_cleaned['Fighting_Style'] = df_cleaned.apply(classify_fighter, axis=1)

# Display count of each fighting style
print(df_cleaned['Fighting_Style'].value_counts())
```

```python
# Group by Fighting Style and calculate average wins and losses
style_comparison = df_cleaned.groupby('Fighting_Style')[['wins', 'losses']].mean().reset_index()


#Visualizinng comparison
plt.figure(figsize=(8,6))
sns.barplot(data=style_comparison, x='Fighting_Style', y='wins', palette='viridis')
plt.title('Average Wins by Fighting Style')
plt.xlabel('Fighting Style')
plt.ylabel('Average Wins')
plt.show()

plt.figure(figsize=(8,6))
sns.barplot(data=style_comparison, x='Fighting_Style', y='losses', palette='magma')
plt.title('Average Losses by Fighting Style')
plt.xlabel('Fighting Style')
plt.ylabel('Average Losses')
plt.show()
```


**STRIKERS**
*   Strikers have the lowest average losses
*   Have the highest average wins and suggestt that fighters are generally more successful in terms of win and loss ratio.

*GRAPPLERS*


*   Have the highest average losses
*   Still, they have a solid wins, indicating that their strategy works but also have higher risk

MIXED STYLE FIGHTERS


*  Have a middle ground and have balanced wins and losses suggesting both offensive and defensive advantages.



```python
# Calculate win percentage

style_comparison['win_percentage'] = (style_comparison['wins'] / (style_comparison['wins'] + style_comparison['losses'])) * 100
```

```python
# Group by fighting style and calculate average win percentage
win_percentage_comparison  = style_comparison.groupby('Fighting_Style')['win_percentage'].mean().reset_index()

# Visualize win percentage comparison
plt.figure(figsize=(8,6))  # Changed to a tuple (width, height)
sns.barplot(data=win_percentage_comparison, x='Fighting_Style', y='win_percentage', palette='coolwarm')
plt.title('Win Percentage by Fighting Style')
plt.xlabel('Fighting Style')
plt.ylabel('Win Percentage(%)')
plt.show()
```

```python
win_percentage_comparison  = style_comparison.groupby('Fighting_Style')['win_percentage'].mean().reset_index()
print(f"Win Percentage Comparison:\n{win_percentage_comparison}")
```


Conclusion
*   Strikers are more efficient and have a higher  win percentage, even though grapplers have a slight edge in total wins
*  Grapplers have more wins but also more losses suggesting they take significant damage over time.
*   So the Mixed Style is balanced and remain close to both categories.

Now we will analyse fighting style vs weight class to check proportion of each syle with each weight class

```python
# Count fighting syle in each weight class

# Use 'df_cleaned' instead of 'df'
weight_class_vs_style = df_cleaned.groupby(["weight_in_kg", "Fighting_Style"]).size().unstack()
```

```python

# Apply classification
def classify_weight(weight):
    if weight <= 61.2:
        return "Flyweight"
    elif weight <= 65.8:
        return "Bantamweight"
    elif weight <= 70.3:
        return "Featherweight"
    elif weight <= 77.1:
        return "Lightweight"
    elif weight <= 83.9:
        return "Welterweight"
    elif weight <= 93:
        return "Middleweight"
    elif weight <= 120.2:
        return "Light Heavyweight"
    else:
        return "Heavyweight"

# Apply classification
df_cleaned["Weight_Class"] = df_cleaned["weight_in_kg"].apply(classify_weight)
```

```python
#Group Data
weight_vs_style = df_cleaned.groupby(["Weight_Class", "Fighting_Style"]).size().unstack() # Use df_cleaned instead of df
```

```python
# Plot Stacked Bar Chart
weight_vs_style.plot(kind='bar', stacked=True, figsize=(14,6))
plt.title("Fighting Style Distribution Across Weight Classes")
plt.xlabel("Weight Class (kg)")
plt.ylabel("Nmber of Fighters")
plt.legend(title="Fighting Style")
plt.xticks(rotation=65)

plt.show()
```
